### Diagramas de cortante y momento 


**Ejemplo 01.** Dibuje los diagramas de cortante y momento para la viga mostrada en la figura.

![](img/ejemplo_01.png)

Por simetría podemos ver que ambas reacciones son de igual magnitud y equivalente a la mitad de la carga aplicada, es decir:

$R_A = R_B = 7\,kN$

Para resolver utilizando SymPy, primero importamos los módulos a utilizar:

```python
from __future__ import division
from sympy import *
from sympy.plotting import *
from sympy.functions.special.singularity_functions import *
init_printing(use_latex="mathjax")
x = symbols("x")
```

Estableciendo la función de carga e integrando para obtener las expresiones para el cortante y momento flexionante:

```python
P = 14e3
RA = P/2
RB = P/2
w = -RA*SingularityFunction(x,0,-1) + P*SingularityFunction(x,1.5,-1) - RB*SingularityFunction(x,3,-1)
V = integrate(-w,x)
M = integrate(V,x)
```

Graficando el diagrama de cortante:

```python
plot(V, (x,0,3), xlabel="L (m)", ylabel="V (N)");
```

Graficando el diagrama de momentos:

```python
plot(M, (x,0,3), xlabel="L (m)", ylabel="M (N*m)");
```
