## Diagramas de cortante y momento

Draw the shear and bending-moment diagrams for problem, 
the beam and loading shown, and determine the maximum absolute
value (a) of the shear, (b) of the bending moment.

### Secci贸n AC

Cortante:

$ -8 - 12x - V = 0 \,\,\,\, \rightarrow \,\,\,\, V = -12x-8 $

Momento:

$ 8x + 12x(\frac{x}{2}) + M = 0  \,\,\,\, \rightarrow \,\,\,\,  M = -6x^2 - 8x$


### Secci贸n CB

Cortante:

$ -8-24+72-V =0 \,\,\,\, \rightarrow \,\,\,\, V=40 $

Momento:

$ 8x + 24(x-1) - 72(x-2) + M = 0 \,\,\,\, \rightarrow \,\,\,\, M = 40x-120 $


```python
from __future__ import division
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(0,2) # Primera secci贸n 
x2 = np.linspace(2,3) # Segunda secci贸n
V1 = -12*x1-8
M1 = -6*x1**2-8*x1
V2 = np.ones_like(x2)*40
M2 = 40*x2-120
x = np.concatenate((x1,x2))
V = np.concatenate((V1,V2))
M = np.concatenate((M1,M2))
```

```python
# Diagrama de cortante
plt.fill_between(x, V, edgecolor="k", lw=2)
plt.xlabel("m")
plt.ylabel("kN")
plt.grid(ls=":")
```

```python
plt.fill_between(x, M, edgecolor="k", lw=2)
plt.xlabel("m")
plt.ylabel(r"kN$\cdot$m")
plt.grid(ls=":")
```
