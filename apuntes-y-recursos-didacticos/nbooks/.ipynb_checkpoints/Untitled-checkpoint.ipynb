```python
from sympy import *
from sympy.plotting import plot
from sympy.physics.continuum_mechanics.beam import Beam
init_printing(use_latex="mathjax")
```

```python
E, I = symbols('E, I')
#E,I,P,a,L = symbols('E,I,P,a,L')
R1, M1 = symbols('R1, M1')
L = 144.
E, I = 29e6, 100
b = Beam(L, E, I)
b.apply_load(R1, L, -1)
b.apply_load(M1, L, -2)
b.apply_load(-15e3, 8*12, -1)
b.apply_load(-2000/12, 0, 0, end=4*12)
b.bc_deflection = [(L, 0),]
b.solve_for_reaction_loads(R1,M1)
```

```python
df = b.deflection()
ss = b.shear_force()
mm = b.bending_moment()
b.reaction_loads
```

```python
plot(ss, ("x",0,L))
```

```python
plot(mm, ("x",0,L))
```
