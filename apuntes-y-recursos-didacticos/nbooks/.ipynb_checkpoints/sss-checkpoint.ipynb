## Diagramas de cortante y momento

Draw the shear and bending-moment diagrams for problem, 
the beam and loading shown, and determine the maximum absolute
value (a) of the shear, (b) of the bending moment.

![](img/5-9-Beer.png)

```python
from __future__ import division
from sympy import *
from sympy.plotting import plot
from sympy.physics.continuum_mechanics.beam import Beam
init_printing(use_latex="mathjax")
```

```python
E, I = symbols('E, I')
RA, RC = symbols('Ra, Rc')
L = 3
#E, I = 29e6, 100
b = Beam(L, E, I)
b.apply_load(RA, 0, -1)
b.apply_load(RC, 2, -1)
b.apply_load(-40e3, 3, -1)
b.apply_load(-12e3, 0, 0, end=2)
b.bc_deflection = [(0, 0),(2,0)]
b.solve_for_reaction_loads(RA,RC)
```

```python
S = b.shear_force()
M = b.bending_moment()
```

```python
plot(S, ("x",0,L))
```

```python
plot(M, ("x",0,L))
```
